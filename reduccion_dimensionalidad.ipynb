## PARTE 2 — CHI-CUADRADO

```python
#chi2: calcula el estadístico Chi²
#SelectKBest: selecciona las mejores variables automáticamente

from sklearn.feature_selection import chi2
from sklearn.feature_selection import SelectKBest
```

```python
#X: variables independientes
#y: variable objetivo

#En machine learning siempre separamos:
# X → lo que usamos para predecir
# y → lo que queremos predecir

X = data[["uso_descuento","visitas_web_mes","compras_previas","satisfaccion"]]
y = data["abandono_cliente"]
```

```python
#chi_scores: qué tan importante es cada variable
#p_values: significancia estadística

# internamente se construye la tabla de contingencia y Calcula frecuencias esperadas
chi_scores, p_values = chi2(X, y)


resultado = pd.DataFrame({
    "variable": X.columns,
    "chi2_score": chi_scores,
    "p_value": p_values
})

resultado.sort_values(by="chi2_score", ascending=False)

#Chi² nos dice qué variables tienen mayor relación con la variable objetivo.
#Entre más alto el valor, más importante es la variable.
```

## Actividad 3

1. ¿Qué variable tiene mayor Chi²?

✏️ 8.853659

2. ¿Qué variable tiene menor Chi²?

✏️ 1.777778

3. ¿Qué significa un valor Chi² alto?

✏️ Un valor Chi² alto significa que existe una mayor asociación o dependencia entre esa variable y la variable objetivo abandono_cliente. En otras palabras, indica que esa variable aporta más información para diferenciar entre clientes que abandonan y clientes que no abandonan.

```python
#Aquí dejamos que el algoritmo decida cuáles son las variables más importantes.

selector = SelectKBest(score_func=chi2, k=2) #Selecciona automáticamente las 2 mejores variables

X_new = selector.fit_transform(X, y) #Aplica el método y Reduce el dataset

mask = selector.get_support() #Devuelve qué variables fueron seleccionadas (True/False)

selected_features = X.columns[mask] #Muestra los nombres de las variables seleccionadas
selected_features
```

4. ¿Cuál es la diferencia principal entre correlación y Chi²?

✏️ La correlación mide la relación lineal entre variables numéricas y además indica si esa relación es positiva o negativa. En cambio, Chi² mide el grado de asociación entre una variable y la variable objetivo, pero no indica dirección, sino qué tan importante o dependiente es esa relación.
