# Min-Max Scaling y Estandarización (Z-Score)

```python
# =========================================
# OBJETIVO
# =========================================
# Comprender cómo funcionan las transformaciones de escalamiento
# y cómo afectan los datos en análisis y modelos.
```

```python
# =========================================
# 📘 PARTE 1: CONCEPTO TEÓRICO
# =========================================
```

```python
# Min-Max Scaling:
# Lleva los datos a un rango entre 0 y 1
# Fórmula:
# X' = (X - min) / (max - min)

# Z-Score:
# Centra los datos en 0 usando la media
# Fórmula:
# Z = (X - media) / desviación estándar
```

```python
# =========================================
# 📊 PARTE 2: CASO DE USO
# =========================================

# Supongamos que tenemos datos de clientes
# Problema: ingresos domina sobre edad
```

```python
import pandas as pd

# Dataset base

data = pd.DataFrame({
    'edad': [20, 25, 30, 40, 50],
    'ingreso': [1000000, 1500000, 3000000, 7000000, 10000000]
})

print("\n📊 Datos originales:\n")
print(data)
```

```python
# =========================================
# 🔹 PARTE 3: MIN-MAX SCALING
# =========================================

from sklearn.preprocessing import MinMaxScaler

# Creamos el escalador
scaler = MinMaxScaler()

# Aplicamos transformación
minmax_data = scaler.fit_transform(data)

# Convertimos a DataFrame
minmax_df = pd.DataFrame(minmax_data, columns=data.columns)

print("\n🔹 Datos con Min-Max Scaling:\n")
print(minmax_df)

# 💡 Observa:
# - Todos los valores están entre 0 y 1
# - Se mantiene la proporción
```

```python
# =========================================
# 🔹 PARTE 4: Z-SCORE (ESTANDARIZACIÓN)
# =========================================

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

zscore_data = scaler.fit_transform(data)

zscore_df = pd.DataFrame(zscore_data, columns=data.columns)

print("\n🔹 Datos con Z-Score:\n")
print(zscore_df)

# 💡 Observa:
# - Valores alrededor de 0
# - Negativos = debajo del promedio
# - Positivos = encima del promedio
```

```python
# =========================================
# 🧪 PARTE 5: VISUALIZACIÓN
# =========================================

#“Comparar cómo cambian los datos antes y después de aplicar transformaciones”

import matplotlib.pyplot as plt

# Datos originales
plt.scatter(data['edad'], data['ingreso'])
plt.title("Datos Originales")
plt.xlabel("Edad")
plt.ylabel("Ingreso")
plt.show()

# MinMax
plt.scatter(minmax_df['edad'], minmax_df['ingreso'])
plt.title("Min-Max Scaling")
plt.xlabel("Edad")
plt.ylabel("Ingreso")
plt.show()

# Z-score
plt.scatter(zscore_df['edad'], zscore_df['ingreso'])
plt.title("Z-Score")
plt.xlabel("Edad")
plt.ylabel("Ingreso")
plt.show()
```

```python
# =========================================
# PARTE 6: ACTIVIDAD
# =========================================

# Ejercicio 1:
# Cambia los valores de ingreso agregando un valor extremo (outlier)
# Ejemplo: 50000000

# Pregunta:
# ¿Qué pasa con Min-Max?
# ¿Qué pasa con Z-score?

# =========================================
# Responde: Escribe las conclusiones
```

```python
# =========================================
# PARTE 6 - EJERCICIO 1
# =========================================

# Nuevo dataset con outlier
data_outlier = pd.DataFrame({
    'edad': [20, 25, 30, 40, 50],
    'ingreso': [1000000, 1500000, 3000000, 7000000, 50000000]
})

print("\n📊 Datos con outlier:\n")
print(data_outlier)

# Min-Max Scaling
scaler_minmax = MinMaxScaler()
minmax_outlier = scaler_minmax.fit_transform(data_outlier)
minmax_outlier_df = pd.DataFrame(minmax_outlier, columns=data_outlier.columns)

print("\n🔹 Datos con Min-Max y outlier:\n")
print(minmax_outlier_df)

# Z-Score
scaler_zscore = StandardScaler()
zscore_outlier = scaler_zscore.fit_transform(data_outlier)
zscore_outlier_df = pd.DataFrame(zscore_outlier, columns=data_outlier.columns)

print("\n🔹 Datos con Z-Score y outlier:\n")
print(zscore_outlier_df)

# Visualizaciones
plt.scatter(data_outlier['edad'], data_outlier['ingreso'])
plt.title("Datos con Outlier")
plt.xlabel("Edad")
plt.ylabel("Ingreso")
plt.show()

plt.scatter(minmax_outlier_df['edad'], minmax_outlier_df['ingreso'])
plt.title("Min-Max con Outlier")
plt.xlabel("Edad")
plt.ylabel("Ingreso")
plt.show()

plt.scatter(zscore_outlier_df['edad'], zscore_outlier_df['ingreso'])
plt.title("Z-Score con Outlier")
plt.xlabel("Edad")
plt.ylabel("Ingreso")
plt.show()
```

**Respuestas de la Parte 6 - Ejercicio 1**

Al agregar el valor extremo 50000000 en la variable ingreso, se observa que la distribución de los datos cambia de manera importante.

En los datos originales con outlier, el último punto queda muy alejado de los demás. Esto muestra claramente que existe un valor atípico que domina la escala vertical del gráfico. Como consecuencia, los otros cuatro puntos quedan visualmente agrupados en la parte inferior.

Cuando se aplica Min-Max Scaling, todos los valores quedan entre 0 y 1. Sin embargo, este método resulta muy sensible al valor extremo. En la columna ingreso, el dato atípico toma el valor 1.0, mientras que los demás quedan muy cercanos a cero (0.010204, 0.040816, 0.122449). Esto indica que el outlier comprime fuertemente los datos normales, reduciendo su separación relativa y haciendo que parezcan mucho más pequeños de lo que realmente eran entre sí.

Con la transformación Z-Score, también se evidencia el efecto del outlier, pero de una forma distinta. El valor extremo queda con un puntaje de 1.987496, mientras que los demás quedan entre aproximadamente -0.61 y -0.29. En este caso, los datos ya no están forzados al intervalo de 0 a 1, sino que se expresan respecto a la media y la desviación estándar. Aunque el valor extremo sigue afectando el resultado, los demás datos conservan una dispersión un poco más interpretable que con Min-Max.

Conclusión

El experimento muestra que Min-Max Scaling es más sensible a los outliers, porque un valor extremo modifica todo el rango y aplasta los demás datos cerca de cero. En cambio, Z-Score también se ve afectado, pero conserva mejor la estructura relativa de los datos. Por tanto, cuando existen valores atípicos, Z-Score suele ser más adecuado que Min-Max Scaling.

```python
# =========================================
# Ejercicio 2:
# Elimina la columna ingreso y prueba solo con edad

# Pregunta:
# ¿Tiene sentido escalar una sola variable?

# =========================================
# Responde: Escribe las conclusiones
```

```python
# =========================================
# PARTE 6 - EJERCICIO 2
# =========================================

# Dataset solo con la variable edad
data_edad = data[['edad']]

print("\n📊 Datos solo con edad:\n")
print(data_edad)

# Min-Max Scaling
scaler_minmax = MinMaxScaler()
minmax_edad = scaler_minmax.fit_transform(data_edad)
minmax_edad_df = pd.DataFrame(minmax_edad, columns=data_edad.columns)

print("\n🔹 Edad con Min-Max Scaling:\n")
print(minmax_edad_df)

# Z-Score
scaler_zscore = StandardScaler()
zscore_edad = scaler_zscore.fit_transform(data_edad)
zscore_edad_df = pd.DataFrame(zscore_edad, columns=data_edad.columns)

print("\n🔹 Edad con Z-Score:\n")
print(zscore_edad_df)

# Visualización datos originales
plt.scatter(range(len(data_edad)), data_edad['edad'])
plt.title("Edad original")
plt.xlabel("Índice")
plt.ylabel("Edad")
plt.show()

# Visualización Min-Max
plt.scatter(range(len(minmax_edad_df)), minmax_edad_df['edad'])
plt.title("Edad con Min-Max Scaling")
plt.xlabel("Índice")
plt.ylabel("Edad")
plt.show()

# Visualización Z-Score
plt.scatter(range(len(zscore_edad_df)), zscore_edad_df['edad'])
plt.title("Edad con Z-Score")
plt.xlabel("Índice")
plt.ylabel("Edad")
plt.show()
```

**Parte 6 – Ejercicio 2**

En este ejercicio se eliminó la columna ingreso y se trabajó únicamente con la variable edad. Al aplicar Min-Max Scaling, los valores se transformaron al intervalo entre 0 y 1, quedando 20 como 0.0 y 50 como 1.0, con los demás valores distribuidos proporcionalmente entre esos extremos. Al aplicar Z-Score, los datos quedaron centrados alrededor de 0: las edades menores que el promedio tomaron valores negativos y las mayores, valores positivos.

Sí tiene sentido escalar una sola variable, porque el procedimiento sigue siendo matemáticamente válido y los resultados muestran claramente la transformación de la escala original. Sin embargo, desde el punto de vista práctico, su utilidad depende del objetivo del análisis. En este caso, como solo se trabaja con una variable y sus valores ya son fáciles de interpretar, el escalado no aporta una ventaja muy grande para la visualización o comprensión directa. Aun así, sí puede ser útil si esa variable se va a usar posteriormente en un modelo de machine learning, en algoritmos sensibles a la escala o en comparaciones con otras variables transformadas.

Conclusión

Escalar una sola variable sí tiene sentido, pero no siempre es necesario. En este ejercicio fue útil para observar cómo cambian los datos con Min-Max y Z-Score, aunque en un análisis simple de una sola columna la interpretación original de la edad sigue siendo más directa.

```python
# =========================================
# Ejercicio 3:
# Crea una nueva variable llamada "gasto" y aplica ambas transformaciones

# Pregunta:
# ¿Cuál método funciona mejor?

# =========================================
# Responde: Escribe las conclusiones
```

```python
# =========================================
# PARTE 6 - EJERCICIO 3
# =========================================


data_gasto = pd.DataFrame({
    'edad': [20, 25, 30, 40, 50],
    'ingreso': [1000000, 1500000, 3000000, 7000000, 10000000],
    'gasto': [500000, 700000, 1200000, 2500000, 4000000]
})

print("\n📊 Datos con nueva variable 'gasto':\n")
print(data_gasto)

# Min-Max Scaling
scaler_minmax = MinMaxScaler()
minmax_gasto = scaler_minmax.fit_transform(data_gasto)
minmax_gasto_df = pd.DataFrame(minmax_gasto, columns=data_gasto.columns)

print("\n🔹 Datos con Min-Max Scaling:\n")
print(minmax_gasto_df)

# Z-Score
scaler_zscore = StandardScaler()
zscore_gasto = scaler_zscore.fit_transform(data_gasto)
zscore_gasto_df = pd.DataFrame(zscore_gasto, columns=data_gasto.columns)

print("\n🔹 Datos con Z-Score:\n")
print(zscore_gasto_df)

# Visualización: edad vs ingreso
plt.scatter(data_gasto['edad'], data_gasto['ingreso'])
plt.title("Datos originales: Edad vs Ingreso")
plt.xlabel("Edad")
plt.ylabel("Ingreso")
plt.show()

plt.scatter(minmax_gasto_df['edad'], minmax_gasto_df['ingreso'])
plt.title("Min-Max: Edad vs Ingreso")
plt.xlabel("Edad")
plt.ylabel("Ingreso")
plt.show()

plt.scatter(zscore_gasto_df['edad'], zscore_gasto_df['ingreso'])
plt.title("Z-Score: Edad vs Ingreso")
plt.xlabel("Edad")
plt.ylabel("Ingreso")
plt.show()

# Visualización: edad vs gasto
plt.scatter(data_gasto['edad'], data_gasto['gasto'])
plt.title("Datos originales: Edad vs Gasto")
plt.xlabel("Edad")
plt.ylabel("Gasto")
plt.show()

plt.scatter(minmax_gasto_df['edad'], minmax_gasto_df['gasto'])
plt.title("Min-Max: Edad vs Gasto")
plt.xlabel("Edad")
plt.ylabel("Gasto")
plt.show()

plt.scatter(zscore_gasto_df['edad'], zscore_gasto_df['gasto'])
plt.title("Z-Score: Edad vs Gasto")
plt.xlabel("Edad")
plt.ylabel("Gasto")
plt.show()
```

**Respuestas al Ejercicio 3**

En este ejercicio se agregó la variable gasto y se aplicaron las transformaciones Min-Max Scaling y Z-Score. Los resultados muestran que ambas funcionan bien, ya que conservan el orden de los datos y permiten comparar mejor las variables edad, ingreso y gasto, que originalmente estaban en escalas muy diferentes.
Con Min-Max, todos los valores quedaron entre 0 y 1, lo que hace más fácil la comparación visual entre variables. Con Z-Score, los datos quedaron centrados alrededor de 0, con valores negativos por debajo del promedio y positivos por encima.
En este caso, Min-Max funciona mejor para interpretar y comparar los datos de forma sencilla, mientras que Z-Score también es útil, pero su interpretación es un poco menos directa. Como no se observan valores atípicos fuertes, ambos métodos son adecuados.

```python
# =========================================
# CONCLUSIÓN
# =========================================

# Min-Max:
# ✔ Útil cuando no hay outliers

# Z-Score:
# ✔ Mejor cuando hay valores extremos

# IDEA CLAVE:
# La transformación depende del tipo de datos
```

```python
#PASO 1
%%writefile app.py
# =========================================
# 🚀 APP STREAMLIT: ESCALAMIENTO INTERACTIVO
# =========================================
# Ejecutar con:
# streamlit run app.py

import streamlit as st
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler
import matplotlib.pyplot as plt

st.title("Escalamiento de Datos Interactivo")

# Crear dataset simple
st.sidebar.header("Configuración de datos")
n = st.sidebar.slider("Número de datos", 5, 100, 20)

np.random.seed(42)
data = pd.DataFrame({
    'edad': np.random.randint(18, 60, n),
    'ingreso': np.random.randint(1000000, 10000000, n)
})

st.subheader("📊 Datos originales")
st.write(data)

# Selección de transformación
option = st.selectbox("Selecciona transformación:", ["Min-Max", "Z-Score"])

if option == "Min-Max":
    scaler = MinMaxScaler()
    transformed = scaler.fit_transform(data)
    st.success("Aplicando Min-Max Scaling")
else:
    scaler = StandardScaler()
    transformed = scaler.fit_transform(data)
    st.success("Aplicando Z-Score")

transformed_df = pd.DataFrame(transformed, columns=data.columns)

st.subheader("📈 Datos transformados")
st.write(transformed_df)

# Visualización
fig, ax = plt.subplots()
ax.scatter(transformed_df['edad'], transformed_df['ingreso'])
ax.set_title("Visualización de datos transformados")
st.pyplot(fig)


# Visualización 2
plt.scatter(data['edad'], data['ingreso'])
plt.title("Datos originales")
plt.show()

st.write("\n🎯 Fin del laboratorio interactivo")
```
