```python
import pandas as pd
import numpy as np

# Cargar archivo
ruta = "/content/clientes_dataset.xlsx"   # en Colab sube aquí el archivo
df = pd.read_excel(ruta, sheet_name="Hoja1")

print("Dimensión del dataset:", df.shape)
print("\nColumnas:")
print(df.columns.tolist())
print("\nTipos de datos:")
print(df.dtypes)
print("\nPrimeras filas:")
display(df.head())
```

**Paso 3 — Identificar correlación con la variable objetivo**

```python
# Correlación de Pearson
corr = df.corr(numeric_only=True)

print("Matriz de correlación:")
display(corr.round(4))
```

```python
# Variable objetivo
objetivo = "abandono_cliente"

# Variables que aparecen en la tabla
variables_tabla = [
    "edad",
    "ingreso_mensual",
    "visitas_web_mes",
    "tiempo_sitio_min",
    "paginas_vistas",
    "clics_email",
    "compras_previas",
    "uso_descuento",
    "satisfaccion"
]

tabla_taller = pd.DataFrame({
    "Variable": variables_tabla,
    "Correlación con abandono_cliente": [corr.loc[var, objetivo] for var in variables_tabla],
    "Tipo de relación (positiva/negativa)": [
        "positiva" if corr.loc[var, objetivo] > 0 else "negativa"
        for var in variables_tabla
    ]
})

tabla_taller["Correlación con abandono_cliente"] = tabla_taller["Correlación con abandono_cliente"].round(4)
display(tabla_taller)
```

**1. ¿Qué significa una correlación negativa en este contexto?**

Una correlación negativa significa que cuando una variable aumenta, la variable abandono_cliente tiende a disminuir. En este contexto, indica una relación inversa con el abandono. Por ejemplo, si aumenta la satisfacción o el ingreso mensual, la tendencia al abandono del cliente disminuye.

**2. Explica la relación entre satisfaccion y abandono_cliente.**

La relación entre satisfaccion y abandono_cliente es negativa y fuerte, con un valor de -0.8750. Esto indica que, a mayor satisfacción del cliente, menor es la probabilidad de abandono. En otras palabras, los clientes más satisfechos tienden a permanecer, mientras que los menos satisfechos presentan mayor tendencia a abandonar.

**3. Explica por qué una variable con correlación cercana a 0 no es útil.**

Una variable con correlación cercana a 0 no es muy útil en este análisis porque no muestra una relación lineal clara con la variable objetivo. Eso significa que su cambio no ayuda a explicar de forma directa el comportamiento de abandono_cliente, por lo que aporta poca información para la selección de atributos usando correlación de Pearson.

**4. ¿Por qué no debemos incluir variables altamente correlacionadas entre sí?**

No se deben incluir variables altamente correlacionadas entre sí porque contienen información muy parecida o redundante. Esto no mejora el análisis y puede hacer que el modelo sea más complejo sin aportar valor adicional. Por eso, en reducción de dimensionalidad conviene quedarse solo con una de las variables redundantes o con la más relevante frente a la variable objetivo.
